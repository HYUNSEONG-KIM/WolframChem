```python
def get_compound_description(parameters, parametername="Name", as_dataframe=False, remainfirst=False, CIDType = False, Method=False):
    if isinstance(parameters,str):
            parameters = [parameters]
    searchdict = {parametername:parameters}
        
    if isinstance(Method,str):
            searchdict["Method"] = Method
    return wc.pubchem("CompoundDescription",searchdict)

    
```
